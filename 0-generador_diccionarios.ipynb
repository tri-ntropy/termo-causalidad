# Prueba de diccionarios: tamaños de ventana $m = 3$ y $m = 4$

### Importación de bibliotecas y módulos

```python
import numpy as np
import modulo_ets as mets
import matplotlib.pyplot as plt
import seaborn as sns
```

### Opciones de graficación
Estética de las gráficas

```python
sns.set_style("darkgrid")
sns.set_context("paper", font_scale = 1.5)
```

## Función para imprimir las series y diccionarios de símbolos

```python
def imprimir(serie, ventana):
    serie_simbolos = mets.simbolizar(serie, m = ventana)
    dicc_simbolos = np.sort(np.array(list(set(serie_simbolos))))
    print("Tamaño de la serie: \n", serie_simbolos.size)
    print("Serie: \n", serie_simbolos)
    print("Tamaño del diccionario: \n", dicc_simbolos.size)
    print("Diccionario: \n", dicc_simbolos)
```

## Dominio temporal

El tiempo en el que se pretende realizar la "simulación"

```python
n = 83
t = np.linspace(0, 1, n)
```

## Simulación

```python
# Simulación de variable continua
a = np.random.normal(0, 0.4, size = n)
a[a > 0.8] = 1.0
a[a < -0.8] = -1.0
# Simulación de una variable discreta: dado de seis caras
b = np.random.randint(1, 7, size = n)
b[16: 20] = 4
b[26: 30] = np.linspace(2, 5, 4)
b[36: 40] = np.linspace(5, 2, 4)
```

## Gráfica de la variable continua

```python
plt.figure(figsize = (20, 5))
plt.ylabel("$\mathbf{X}$")
plt.xlabel("$t$")
plt.plot(t, a, "o--")
plt.savefig("graficas/diccionarios/ruido_blanco.pdf", bbox_inches = "tight")
plt.show()
```

## Simulación del dado

```python
plt.figure(figsize = (20, 5))
plt.ylabel("$\mathbf{Y}$")
plt.xlabel("$t$")
plt.plot(t, b, "o--")
plt.savefig("graficas/diccionarios/dado_seiscaras.pdf", bbox_inches = "tight")
plt.show()
```

## Serie y diccionario de simbolos de la variable continua con $m = 3$

```python
imprimir(a, 3)
```

## Serie y diccionario de simbolos de la variable continua con $m = 4$

```python
imprimir(a, 4)
```

## Serie y diccionario de simbolos del dado con $m = 3$

```python
imprimir(b, 3)
```

## Serie y diccionario de simbolos del dado con $m = 4$

```python
imprimir(b, 4)
```
