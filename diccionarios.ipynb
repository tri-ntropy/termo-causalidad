### Importación de bibliotecas y módulos

```python
import numpy as np
import modulo_ets as ets
import matplotlib.pyplot as plt
import seaborn as sns
```

### Opciones de graficación
Estética de las gráficas

```python
sns.set_style("darkgrid")
sns.set_context("paper", font_scale = 1.5)
```

```python
def imprimir(serie, ventana):
    serie_simbolos = ets.simbolizar(serie, m = ventana)
    dicc_simbolos = np.sort(np.array(list(set(serie_simbolos))))
    print("Tamaño de la serie: \n", serie_simbolos.size)
    print("Serie: \n", serie_simbolos)
    print("Tamaño del diccionario: \n", dicc_simbolos.size)
    print("Diccionario: \n", dicc_simbolos)
```

```python
n = 63
a = np.random.normal(0, 0.4, size = n)
b = np.random.randint(1, 7, size = n)
b[16: 20] = 6
b[26: 30] = np.linspace(2, 5, 4)
b[36: 40] = np.linspace(5, 2, 4)
t = np.linspace(0, 1, n)
```

```python
plt.figure(figsize = (20, 5))
plt.ylabel("$\mathbf{X}$")
plt.xlabel("$t$")
plt.plot(t, a, "o--")
plt.savefig("grafica_diccionarios/ruido_blanco.pdf", bbox_inches = "tight")
plt.show()
```

```python
plt.figure(figsize = (20, 5))
plt.ylabel("$\mathbf{Y}$")
plt.xlabel("$t$")
plt.plot(t, b, "o--")
plt.savefig("grafica_diccionarios/dado_seiscaras.pdf", bbox_inches = "tight")
plt.show()
```

```python
imprimir(a, 3)
```
